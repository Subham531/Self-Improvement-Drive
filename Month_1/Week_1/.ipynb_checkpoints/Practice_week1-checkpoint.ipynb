```python
import pandas as pd
```

```python
df=pd.read_csv('Auto_Sales_data.csv')
df
```

# Filtering rows based on quantity

```python
df = df[df['QUANTITYORDERED']>20]
df =  df[df['DEALSIZE'] == 'Medium']
df.shape
```

# Feature selection based on importance

```python
features = ['ORDERNUMBER','QUANTITYORDERED','PRICEEACH','SALES','ORDERDATE','PRODUCTCODE','COUNTRY','DEALSIZE']
# df = df[features]
df.shape
```

```python
df.groupby('COUNTRY')['DEALSIZE'].count()
```

```python
df.groupby('COUNTRY')['SALES'].sum()
```

```python
df.groupby('COUNTRY')['QUANTITYORDERED'].mean()
```

```python
oldest_date,latest_date = df['ORDERDATE'].min(),df['ORDERDATE'].max()
oldest_date,latest_date
```

```python
df.groupby('COUNTRY').agg({'SALES':['count','sum','mean']})
```

```python
df_feature = df[features]
df_feature
```

```python
df
```

```python
df = df[['PRODUCTLINE','STATUS','MSRP','CITY','COUNTRY']]
```

# Inner Merge

```python
pd.merge(df,df_feature,on='COUNTRY')
```

# Left Merge

```python
new_row = pd.DataFrame({'COUNTRY':['INDIA']})
df = pd.concat([df,new_row])
df
```

```python
pd.merge(df,df_feature,how='left',on='COUNTRY')
```

# Right Merge

```python
pd.merge(df,df_feature,how='right',on='COUNTRY')
```

# Outer Merge

```python
pd.merge(df,df_feature,how='outer',on='COUNTRY')
```

# JOINING DFs

```python
df.join(df_feature, lsuffix='_left', rsuffix='_right')
```

# CONCAT DFs

```python
pd.concat([df,df_feature])
```

# Pivot table

```python
pivot = pd.pivot_table(df,index='COUNTRY',columns='PRODUCTLINE',values='SALES',aggfunc='mean')
pivot
```
