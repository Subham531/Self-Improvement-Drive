```python
import pandas as pd
```

```python
df = pd.read_csv('tested.csv')
df
```

# checking for null values

```python
df.isna().sum()
```

# filling with mean

```python
df['Age'].isna().sum() # 86

df['Age'] = df['Age'].fillna(df['Age'].mean())
df.isna().sum()
```

# filling with median

```python
df['Age'] = df['Age'].fillna(df[['Age']].median())
df.isna().sum()
```

# filling with ffill

```python
df['Age'] = df['Age'].fillna(method='ffill') # fills the nan value as same as the last obtaining value
df.isna().sum()
```

# filling with bfill

```python
df['Age'] = df['Age'].fillna(method='bfill') # fills the nan value as same as the next obtaining value
df.isna().sum()
```

# filling with KNN-imputer 

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import KNNImputer

scaler = MinMaxScaler()

df_Age_scaled = scaler.fit_transform(df[['Age']])

imputer = KNNImputer(n_neighbors=2)
df_imputer_scaled = imputer.fit_transform(df_Age_scaled)
df_imputer_scaled
```
