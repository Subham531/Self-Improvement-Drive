```python
import pandas as pd
import numpy as np
df = pd.read_csv('online_retail_real_world.csv')
df
```

```python
df.isna().sum()
```

```python
df['Raw_Weight'].dropna(inplace=True)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python


numeric_features = ['UnitPrice']
categorical_features = ['ProductName','Brand']



numeric_Pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])


categorical_Pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('one-hot_encoding', OneHotEncoder(sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
    ('num', numeric_Pipeline,numeric_features),
    ('cat', categorical_Pipeline,categorical_features)
])
```

```python
clean_train_data = preprocessor.fit_transform(df)

print("✅ CLEANED TRAINING DATA (Notice no missing values):")
print(np.round(clean_train_data, 2), "\n")
```
