# Implementing IQR method: 


```python

np.random.seed(42)
df = pd.DataFrame({
    'Integers': np.random.randint(1, 100, size=10),  # 10 random integers from 1 to 99
    'Floats': np.random.rand(10)                     # 10 random floats between 0 and 1
})

Q1 = np.percentile(df['Integers'],25)
Q2 = np.percentile(df['Integers'],50)
Q3 = np.percentile(df['Integers'],75)

Q1,Q2,Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
Lower_bound = Q1 - (1.5 * IQR)
Upper_bound = Q3 + (1.5 * IQR)

outlier = [value for value in df['Integers'] if value > Upper_bound or value <Lower_bound]
outlier
```

# Implementing z-score method

```python
df['z_score'] = (df['Floats'] - df['Floats'].mean())/df['Floats'].std()
outlier = df[(df['z_score'] > 3) | (df['z_score'] < -3)]
outlier
# df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
normal_data = np.random.normal(loc=50, scale=5, size=100)
outliers = np.array([20, 25, 85, 90, 95]) 
df = pd.DataFrame({'Values': np.concatenate([normal_data, outliers])})

plt.figure(figsize=(12,6))

sns.boxplot(y=df['Values'],color='lightblue',)
plt.show()
```
