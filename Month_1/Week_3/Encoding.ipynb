```python
import pandas as pd
import numpy as np
```

```python
data = {
    'City': ['New York', 'London', 'New York', 'Tokyo', 'London'],
    'Education': ['Bachelors', 'PhD', 'Masters', 'Bachelors', 'Masters'],
    'Salary': [70000, 120000, 95000, 65000, 100000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
targeted_encoding = df.groupby('City')['Salary'].mean()
df['City_target'] = df['City'].map(targeted_encoding)
# df.drop(columns={'City_targeted'},inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder,TargetEncoder

# Labelencoding :-
le = LabelEncoder()
df['City_label'] = le.fit_transform(df['City'])
df
```

```python
# OneHotEncoding
ohe = OneHotEncoder(sparse_output=False,drop='first')
encoder_data = ohe.fit_transform(df[['Education']])
```

```python
encoder_data
```

```python
# TargetEncoding
te = TargetEncoder(target_type='continuous',cv=2,random_state=42)
df['City_targeted'] = te.fit_transform(X = df[['City']],y = df['Salary'])
df
```
