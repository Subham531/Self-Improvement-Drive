```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler 
%matplotlib inline
```

```python
np.random.seed(42)
normal = np.random.normal(loc=50,scale=10,size=1000)
outliers = np.array([-50,150,200,185])
data = np.concatenate([normal,outliers]).reshape(-1,1)

df = pd.DataFrame(data,columns=['Original'])
df['Standard_Scaler'] = StandardScaler().fit_transform(df[['Original']])
df['MinMax_Scaler'] = MinMaxScaler().fit_transform(df[['Original']])
df['Robust_Scaler'] = RobustScaler().fit_transform(df[['Original']])
```

```python
df
```

```python
sns.displot(df['Original'],kde=True)
plt.show()
```

```python
sns.displot(df['Standard_Scaler'],kde=True)
plt.show()
```

```python
sns.displot(df['MinMax_Scaler'],kde=True)
plt.show()
```
