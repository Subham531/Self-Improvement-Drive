# Checking for null values and duplicate values

```python
df.drop_duplicates(inplace=True)
df.shape  # No duplicates 
```

```python
df.isna().sum() # Lot of null values out here in age,cabin and fare
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Fare'].fillna(df['Fare'].mode(),inplace=True)
```

```python
df.drop(columns={'Cabin','Name','Embarked','Ticket'},inplace=True)
```
